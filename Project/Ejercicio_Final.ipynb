```python
# importamos las librerias necesarias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# leemos el dataset
data_file='Project\data\datos_pago_microcreditos.csv'
pdl=pd.read_csv(data_file)
```

```python
# remlazamos los valores de la columna payment por 1 y 0 de acuerdo a las reglas del establecidadas
pdl["payment"]=np.where(pdl["payment"]=="Success",1,0)

k=pdl.columns # listamos las columnas del dataset

# modificamos las varialbes categoricas a dummies
for col in k:
    if pdl[col].dtype=='object':
        temp=pd.get_dummies(pdl[col],drop_first=True,prefix=col)
        pdl=pd.concat([pdl,temp],axis=1)
        pdl.drop([col],axis=1,inplace=True)
# eliminamos las filas que tienen valores nulos
pdl.dropna(axis=0,inplace=True)
```

```python
pdl.info() # verificamos que no hay valores nulos y la info de las columnas 
```

```python
pdl.head(10) # mostramos las primeras 10 filas del dataset
```

```python
pdl_train, pdl_test = train_test_split(pdl, test_size = 0.2,random_state=2) # dividimos el dataset en train y test, 80% y 20% respectivamente

#x_train=pdl_train.drop(["payment"],1)
x_train=pdl_train.drop(columns="payment") # variable independiente
y_train=pdl_train["payment"] # variable objetivo

#x_test=pdl_test.drop(["payment"],1)
x_test=pdl_test.drop(columns="payment") # variable independiente
y_test=pdl_test["payment"] # variable objetivo

x_train.reset_index(drop=True,inplace=True) # reiniciamos el index
y_train.reset_index(drop=True,inplace=True) # reiniciamos el index
```

# Pregunta 3:
Usando GridSearchCV, identifica para un clasificador K-NN el numero de vecinos entre 1 y 30 que optimiza el resultado usando como scoring la precision y con 10 folds para la validacion cruzada. 

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

# definimos el rango de vecinos de 1 a 30
param_grid = {'n_neighbors': np.arange(1, 31, 1)}

# eonfiguramos GridSearchCV con 10 folds y el clasificador K-NN
grid_search = GridSearchCV(
    KNeighborsClassifier(),
    param_grid=param_grid,
    scoring='accuracy',  # usamos el scoring de precision
    cv=10  # 10 dobles
)

# entrenamos el modelo
grid_search.fit(x_train, y_train)


print(f'el mejor promedio de accuracy con validacion cruzada es: {grid_search.best_score_:.3f}')
print(f'El mejor parametro de numero de vecinos es: {grid_search.best_params_}')

# evaluamos el modelo en el conjunto de test
print(f'El accuracy del modelo test es: {grid_search.score(x_test, y_test):.3f}')
```

# Pregunta 4:
Obten la matriz de confusion del clasificador optimo anterior. ¬øcomo lees cada uno de los valores?¬øque valor de precission y recall tiene el clasificador optimizado con precission?

```python
# importamos la libreria
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score

#  predicciones del clasificador anterior
y_pred = grid_search.best_estimator_.predict(x_test)

# matriz de confusi√≥n
conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusi√≥n:")
print(conf_matrix)

# precisi√≥n y recall
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print(f'\nPrecisi√≥n del clasificador optimizado anteriormente: {precision:.3f}')
print(f'Recall del clasificador optimizado: {recall:.3f}')

# Reporte de clasificaci√≥n para m√°s detalles
print('\nReporte de Clasificaci√≥n:')
print(classification_report(y_test, y_pred))
```

La matriz de confucion nos muestra que:

* Tenemos 1323 FN - Falsos negativos: Pagos incorrectamente clasificados como 'No pago' 
* Tenemos 915 TP - Verdaderos positivos: Pago correctamente clasificados como 'Pago'

Podemos ver que tenemos buen desempe√±o para identificar los 'No pago' de manera correcta, pero bajo desempe√±o en indentificar los 'Pago' de manera correcta


* En cuanto a la precision tenemos un valor de 0.73 , lo que nos dice que el 73% de los pagos clasificados como 'Pago' son correctas.
* El recall tiene un valor de 0.41 para la clase 'Pago', lo que me indica que el modelo identifica correctamente el 41% de los valores 'Pago'
* Para el valor de F1 tenemos 0.52 para la clase 'Pago', indica un "equilibrio" entre la presicion y el recall
* El accuracy del modelo es de 0.72, estando en un valor no muy malo


Conclusion: Tenemos falencias a la hora de clasificar correctamente los valores 'Pago', se deberia ajustar para asi evitar falsos positivos

# Pregunta 5:
Construye ahora una red neuronal usando MLPClassifier de dos capas. Prueba diferente numero de neuronas por capa. ¬øafecta al resultado?

```python
from sklearn.neural_network import MLPClassifier

# definimos diferentes numero de neuronas por capa
hidden_layer_sizes = [(10, 10), (20, 10), (50, 25), (120,60)]

for layers in hidden_layer_sizes:
    print(f"\nred neuronal con capas ocultas: {layers}")
    
    mlp = MLPClassifier(hidden_layer_sizes=layers, max_iter=1000, random_state=0)
    mlp.fit(x_train, y_train)
    
    # evaluamos el modelo
    y_pred = mlp.predict(x_test)
    accuracy_test = mlp.score(x_test, y_test)
    accuracy_train = mlp.score(x_train, y_train)
    print(f'Accuracy en train: {accuracy_train:.3f}')
    print(f"Accuracy en test: {accuracy_test:.3f}")

    # mostramos la matriz de confusi√≥n y el reporte de clasificaci√≥n
    print("Matriz de Confusi√≥n:")
    print(confusion_matrix(y_test, y_pred))
    print("\nReporte de Clasificaci√≥n:")
    print(classification_report(y_test, y_pred))
```

Al observar los resultados, podemos identificar que a medida que aumentamos el numero de neuronas nuestro modelos va entrando en sobren ajuste, lo que nos indica que la red menos compleja (10,10) dos capas, de 10 neuronas cada una es la mas adecuada, el que mas generaliza:

# Pregunta 6:
GradientBoostingClasifier es uno de los metodos de scikitlearn que mejor resultados suelen producir. Implementa un clasificador utilizando GradientBoostingClasifier considerando 0.05, 0.1, 0.15, 0.2, 0.25, 0.3 como learning rate.¬øque learning rate produce un resultado optimo usando recall?

### 1. Implementando GridSearchCV

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import recall_score
from sklearn.model_selection import GridSearchCV

# valores para el learning_rate
param_grid = {'learning_rate': [0.05, 0.1, 0.15, 0.2, 0.25, 0.3]}

gbc = GradientBoostingClassifier(random_state=0) # con max_depth por defecto y n_estimators tambien debido a que posteriormente
# se optimizara el learning_rate con el metodo de validacion cruzada GridSearchCV

# GridSearchCV para optimizar el recall con 10 folds haciendo validaci√≥n cruzada
grid_search_gbc = GridSearchCV(estimator=gbc,
                               param_grid=param_grid,
                               scoring='recall',
                               cv=10)

grid_search_gbc.fit(x_train, y_train)

# Mostramos el mejor learning_rate basado en recall
print(f"Mejor learning_rate: {grid_search_gbc.best_params_['learning_rate']}")
print(f"Mejor recall en validaci√≥n: {grid_search_gbc.best_score_:.3f}")

# Evaluamos el recall en el conjunto de test y train
y_pred = grid_search_gbc.best_estimator_.predict(x_test)
y_pred_train = grid_search_gbc.best_estimator_.predict(x_train)
print(f'Recall en train es: {recall_score(y_pred_train, y_train):.3f}')
print(f'Recall en test: {recall_score(y_test, y_pred):.3f}')
```

```python
# para responder la ultima pregunta, imprimimos la matriz de confusi√≥n
conf_matrix_gbc = confusion_matrix(y_test, y_pred)
print("Matriz de Confusi√≥n del Gradient Boosting Classifier:")
print(conf_matrix_gbc)
```

### 2. Implementado un bluque sin GridSearchCV
Esto para ver cada recall de learning rate

```python
# Por otro lado me gustaria hacer un bucle para ver el recall de cada learning_rate
results = {}

for lrate in param_grid['learning_rate']:
    gbc = GradientBoostingClassifier(learning_rate=lrate, random_state=0) # con dejar max_depth por defecto y n_estimators
    gbc.fit(x_train, y_train)
    y_pred = gbc.predict(x_test)
    rec = recall_score(y_test, y_pred)
    recall_train = recall_score(y_train, gbc.predict(x_train))
    results[lrate] = rec
    print(f"Con Learning_rate ={lrate:.2f} El Recall en test: {rec:.3f} y en train: {recall_train:.3f}")

best_lrate = max(results, key=results.get)
print(f"\nEl Mejor learning_rate (por recall en test): {best_lrate} con recall={results[best_lrate]:.3f}")
```

El mejor learning rate es 0.2, tiene un score de 0.807 para test y en train un 0.817, es el learning que genera el modelo que mejor generaliza
